```python
import pandas as pd
titanic_train = pd.read_csv('https://raw.githubusercontent.com/training-ml/Files/main/titanic_train.csv')

titanic_train
```

```python
survived_df = titanic_train[ titanic_train['Survived'] ==1 ]
survived_df
```

```python
died_df = titanic_train[titanic_train['Survived']==0]
died_df
```

```python
print('Total = ', len(titanic_train))
```

```python
print('number of passangers who survived =', len(survived_df))
```

```python
print('number of passangers who died =', len(died_df))
```

```python
print('percentage survived =', 1. * len(survived_df) / len(titanic_train) * 100.0, '%')
```

```python
print('percentage died =', 1. * len(died_df) / len(titanic_train) * 100.0, '%')
```
